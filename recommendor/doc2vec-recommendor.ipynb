```python
import pandas as pd
```

```python
restaurants = pd.read_csv('restaurants.csv')
restaurants.head()
```

```python
meta_data = pd.read_excel('List_Of_Restaurants.xlsx')
meta_data.head()
```

```python
restaurants['doc'] = meta_data['About Section'] + meta_data['5_Reviews']
restaurants.iloc[0]['doc']
```

```python
restaurants = restaurants.replace(r'\n',' ', regex=True) 
```

```python
restaurants.iloc[0]['doc']
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

```python
tagged_data = [
    TaggedDocument(
        words=word_tokenize(row['doc'].lower()),
        tags=[str(row['id'])]
    )
    for index, row in restaurants.iterrows()
]
```

```python
tagged_data
```

```python
max_epochs = 100
vec_size = 30
# initial learning rate
alpha = 0.025

model = Doc2Vec(
    vector_size=vec_size,
    alpha=alpha,
    min_alpha=0.00025,
    min_count=1,
    dm=1
)

model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)
    
    model.alpha -= 0.0002
    model.min_alpha = model.alpha

model.save('d2v.model')
print('Model saved')
```

```python
test_data = word_tokenize(restaurants.iloc[0]['doc'].lower())
v1 = model.infer_vector(test_data)
print('V1 infer', v1)
```

```python
similar_doc = model.dv.most_similar([v1])
print(similar_doc)
```

```python
restaurants[restaurants['id'] == int(similar_doc[1][0])]
```
